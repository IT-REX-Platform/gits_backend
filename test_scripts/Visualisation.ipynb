```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots


media_old_df = pd.read_csv('data/media_old.csv')
```

```python
media_old_waiting = media_old_df.loc[(media_old_df['metric_name'] == 'http_req_waiting') | (media_old_df['metric_name'] == 'http_reqs')].reset_index()
media_old_waiting = media_old_waiting[['timestamp', 'metric_name', 'metric_value']]

media_old_http_reqs_grouped = media_old_waiting.loc[media_old_waiting['metric_name'] == 'http_reqs'].groupby('timestamp').sum(numeric_only=True).reset_index().rename(columns={'metric_value':'http_reqs'})

media_old_waiting_grouped = media_old_waiting.loc[media_old_waiting['metric_name'] == 'http_req_waiting'].groupby('timestamp').mean(numeric_only=True).reset_index().rename(columns={'metric_value':'http_req_waiting'})

media_old_vus = media_old_df.loc[media_old_df['metric_name'] == 'vus'].reset_index()
media_old_vus = media_old_vus[['timestamp', 'metric_value']].rename(columns={'metric_value':'vus'})

media_old_http_reqs_grouped['timestamp'] = pd.to_datetime(media_old_http_reqs_grouped['timestamp'], unit='s')
media_old_waiting_grouped['timestamp'] = pd.to_datetime(media_old_waiting_grouped['timestamp'], unit='s')
media_old_vus['timestamp'] = pd.to_datetime(media_old_vus['timestamp'], unit='s')
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
x=media_old_http_reqs_grouped['timestamp'], y=media_old_http_reqs_grouped['http_reqs'], name='Requests per Second'))

fig.add_trace(go.Scatter(
x=media_old_waiting_grouped['timestamp'], y=media_old_waiting_grouped['http_req_waiting'], name='Response Time in Milliseconds'))

fig.add_trace(go.Scatter(
x=media_old_vus['timestamp'], y=media_old_vus['vus'], name='Number of Users'))

fig.update_layout(title_text="Media Service")

fig.show()
```
